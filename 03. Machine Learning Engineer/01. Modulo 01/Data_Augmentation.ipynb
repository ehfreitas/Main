```python
import numpy as np
import cv2 # Importa biblioteca OpenCV
import requests
from PIL import Image
from io import BytesIO
from google.colab.patches import cv2_imshow
```

```python
# Realiza requisição da image e converte para imagem
response = requests.get('https://images.bisnis.com/posts/2017/03/23/639731/katak.jpg')
img = Image.open(BytesIO(response.content))
img = np.array(img)
cv2_imshow(img)
```

```python
# Importa biblioteca de Data Augmentation
from keras.preprocessing.image import ImageDataGenerator

# Cria objeto de data augmentation
aug = ImageDataGenerator(height_shift_range=0.2, width_shift_range=0.2, rotation_range=30, zoom_range=[0.8, 1.0])

# Cria uma nova dimensão
imagem = np.array(img)
imagem = np.expand_dims(imagem, axis=0)

# Realiza data augmentation com imagem passada
imageGen = aug.flow(imagem, batch_size=1)

# Imprime 5 primeiras imagens geradas
for i in range(5):
  next = imageGen.next()
  imgPrint = np.array(next)
  cv2_imshow(imgPrint[0])
```
